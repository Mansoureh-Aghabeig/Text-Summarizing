### First techniuque for Prediction  (More costomized)

```python
# Step 3: Example input text
article = """
Once the training process is complete, we need to evaluate the model and check how well it summarizes the dialogue. While ROUGE and BLEU scores are commonly used for machine translation and summarization, we will use the ROUGE score in this tutorial.

BLEU (Bilingual Evaluation Understudy) Score is a prominent metric extensively utilized in the realm of machine translation tasks. BLEU score is precision-focused. The primary objective of these tasks is to automate the translation of text from one language to another. The BLEU score quantifies the similarity between the text translated by the machine and the reference translations. This measurement is conducted using n-grams, which are contiguous sequences of â€™nâ€™ words.

ROUGE (Recall-Oriented Understudy for Gisting Evaluation) Score is a suite of metrics utilized for assessing automatic text summarization and machine translation. ROUGE score focuses on recall.It compares an auto-generated summary or translation against a reference or set of references. ROUGE scores, ranging from 0 to 1, reflect the similarity between the machine-produced summary and the reference, with higher scores indicating greater similarity.
"""

# Step 4: Add "summarize:" prefix (T5 needs it!)
input_text = "summarize: " + article.strip()

# Step 5: Tokenize the input
inputs = tokenizer([input_text], max_length=512, truncation=True, padding="max_length", return_tensors="pt").to(device)

# Step 6: Generate summary
summary_ids = model.generate(
    input_ids=inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    length_penalty=1,
    num_beams=8,
    max_length=128
)


# Step 7: Decode summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("ğŸ“ Generated Summary:")
print(summary)

```

```python
def convert_examples_to_features(example_batch):
    # Add the "summarize:" prefix required by T5
    input_texts = ["summarize: " + dialog for dialog in example_batch['dialogue']]

    # Tokenize inputs
    input_encodings = tokenizer(
        input_texts,
        max_length=512,              # T5 supports max 512 tokens
        truncation=True,
        padding="max_length"
    )

    # Tokenize targets (summaries)
    target_encodings = tokenizer(
        example_batch['summary'],
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    return {
        'input_ids': input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

âœ… **What is a Data Collator?**

A **data collator** is a function (or object) that:

- Takes a **list of examples** from your dataset (like a batch of input/target pairs),
- Automatically **pads them to the same length** (since inputs and targets vary in length),
- Returns them as a **batch of tensors**, ready for model training.


```python
# Training

from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='t5-small-samsum',  # Updated output directory
    num_train_epochs=1,
    warmup_steps=500,
    per_device_train_batch_size=4,  # You can increase batch size with t5-small
    per_device_eval_batch_size=4,
    weight_decay=0.01,
    logging_steps=10,
    evaluation_strategy='steps',
    eval_steps=500,
    save_steps=int(1e6),  # Convert to int to avoid float issues
    gradient_accumulation_steps=4,  # Adjust based on your GPU memory
    report_to="none"
)
```

```python

trainer = Trainer(model=model, args=trainer_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt["test"],
                  eval_dataset=dataset_samsum_pt["validation"])
```

```python
# Evaluation

def generate_batch_sized_chunks(list_of_elements, batch_size):
    """split the dataset into smaller batches that we can process simultaneously
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]



def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                               batch_size=16, device=device,
                               column_text="dialogue",
                               column_summary="summary"):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches), total=len(article_batches)):

        inputs = tokenizer(article_batch, max_length=1024,  truncation=True,
                        padding="max_length", return_tensors="pt")

        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                         attention_mask=inputs["attention_mask"].to(device),
                         length_penalty=0.8, num_beams=8, max_length=128)
        ''' parameter for length penalty ensures that the model does not generate sequences that are too long. '''

        # Finally, we decode the generated texts,
        # replace the  token, and add the decoded texts with the references to the metric.
        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                clean_up_tokenization_spaces=True)
               for s in summaries]

        decoded_summaries = [d.replace("", " ") for d in decoded_summaries]


        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    #  Finally compute and return the ROUGE scores.
    score = metric.compute()
    return score
```
